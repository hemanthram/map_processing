```python
import requests 
from bs4 import BeautifulSoup 
from csv import writer
```

```python
for j in range(174):
    if(j == 1):
        continue
    URL = "https://www.vcsdata.com/companies_bangalore.php?page="+str(j)+"&city=Bengaluru%20/Bangalore"
    r = requests.get(URL) 

    soup = BeautifulSoup(r.content, 'html5lib') 

    table = soup.findAll('div', attrs = {'class':'innertitle'})#'class':'contentBoxListing','id':'company_list'}) 
    cname = []
    for row in table: 
#         print(row.h3.text)
        cname.append(row.h3.text)
    table = soup.findAll('div', attrs = {'itemprop':"address"})#'class':'contentBoxListing','id':'company_list'}) 
    i = 0
    for row in table: 
        adr = (row.h6.text)
        pin = (row.find('span', attrs = {'itemprop':"postalCode"}).text)
        typ = (row.findAll('div', attrs = {'class':"col-md-6"})[1].h6.text)
        with open('company.csv', 'a+', newline='') as write_obj:
            # Create a writer object from csv module
            csv_writer = writer(write_obj)
            # Add contents of list as last row in the csv file
            try:
                csv_writer.writerow([cname[i],adr,pin,typ])
            except:
                print('Error')
        i += 1
    print(j)
```
